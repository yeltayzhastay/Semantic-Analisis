```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from wordcloud import WordCloud

import warnings
warnings.filterwarnings('ignore')
```

# 1 Version

```python
data = pd.read_json("dataset/News_Category_Dataset_v2.json",lines=True)
data = pd.concat([data[data.category != 'CRIME'][data.short_description != ''][data.authors != ''].sample(3405), data[data.category == 'CRIME'][data.short_description != '']])
# data[['category']].to_csv('dataset/cat.txt', index=False, sep='~')
# data[['short_description']].to_csv('dataset/text.txt', index=False, sep='~')
```

```python
data.info()
```

```python
data.isna().apply(pd.value_counts) #missing value check
```

```python
data.category.nunique()
```

```python
data.category.unique()
```

```python
data.category = data.category.map(lambda x: "WORLDPOST" if x == "THE WORLDPOST" else x)
```

```python
group_count = data['category'].value_counts()
sns.barplot(group_count.index[:10], group_count.values[:10], alpha=0.8)
plt.title('Top 10 category ')
plt.ylabel('Counts', fontsize=12)
plt.xlabel('Category groups', fontsize=12,)
plt.xticks(rotation=45)
plt.show()
```

```python
group_count = data['authors'].value_counts()
sns.barplot(group_count.index[:10], group_count.values[:10], alpha=0.8)
plt.title('Top 10 authors')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Author types', fontsize=12,)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Plotting word cloud
from wordcloud import WordCloud
desc = " ".join(str(des) for des in data['headline'])

wc_desc = WordCloud(background_color='white', max_words=200, width=400, height=400,random_state=10).generate(desc)
plt.figure(figsize=(10,10))
plt.imshow(wc_desc)
plt.title("Word cloud for Headline column")
```

```python
#Plotting word cloud for short description
from wordcloud import WordCloud
desc = " ".join(str(des) for des in data['short_description'])

wc_desc = WordCloud(background_color='white', max_words=200, width=400, height=400,random_state=10).generate(desc)
plt.figure(figsize=(10,10))
plt.imshow(wc_desc)
plt.title("Word cloud for short description column")
```

```python
c = pd.read_csv('dataset/Crime.txt', sep='{}', encoding='utf8')
c['label'] = 1
nc = pd.read_csv('dataset/NotCrime.txt', sep='{}', encoding='utf8')
nc['label'] = 0
```

```python
df = pd.concat([c, nc])
df
```

```python
from wordcloud import WordCloud
desc = " ".join(str(des) for des in df['kz'])

wc_desc = WordCloud(background_color='white', max_words=200, width=400, height=400,random_state=10).generate(desc)
plt.figure(figsize=(10,10))
plt.imshow(wc_desc)
plt.title("Word cloud for kazakh version")
```

# 2 Version

```python
import os
path = 'dataset/Data/'
data_list = []
for i in os.listdir(path):
    for j in os.listdir(path+i+'/'):
        data_list.append([i, open(path+i+'/'+j, "r").read()])
```

```python
df = pd.DataFrame(data_list, columns = ['category', 'text'])
df
```

```python
#Plotting word cloud
from wordcloud import WordCloud
desc = " ".join(str(des) for des in df['text'])

wc_desc = WordCloud(background_color='white', max_words=200, width=400, height=400,random_state=10).generate(desc)
plt.figure(figsize=(10,10))
plt.imshow(wc_desc)
plt.title("Word cloud")
```
